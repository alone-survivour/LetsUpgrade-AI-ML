### **2. Problem Statement and Objective**

**Problem Statement:**<br>

A large company named XYZ, employs, at any given point of time, around 4000 employees. However, every year, around 15% of its employees leave the company and need to be replaced with the talent pool available in the job market. The management believes that this level of attrition (employees leaving, either on their own or because they got fired) is bad for the company, because of the following reasons -

1. The former employeesâ€™ projects get delayed, which makes it difficult to meet timelines, resulting in a reputation loss among consumers and partners.<br>
2. A sizeable department has to be maintained, for the purposes of recruiting new talent. More often than not, the new employees have to be trained for the job and/or given time to acclimatise themselves to the company.<br>

Hence, the management has contracted an HR analytics firm to understand what factors they should focus on, in order to curb attrition. In other words, they want to know what changes they should make to their workplace, in order to get most of their employees to stay. Also, they want to know which of these variables is most important and needs to be addressed right away.


**Objective:**<br>

1. To perform statistical, univariate and correlational analysis on the given data and form inferences accordingly.<br>
2. To find out patterns of features like age, monthly income, distance from home and other features using univariate analysis.<br>
3. To find out if features like age of the employee, monthly income, years with company have any kind of relation with attrition using correlational analysis.

### **5. Performing Analysis using statistical tests**

**5.1 Hypotheses and statistical tests**

1. Hypotheses are potential explanations that can account for our observations of the external world. They usually describe cause-and-effect relationships between a proposed mechanism or process (the cause) and our observations (the effect). 

2. If observations are the â€œwhatâ€ of science, hypotheses are the â€œhowâ€. Whereas
observations are taken from the real world, hypotheses need not be.

3. The hypothesis should be clear and precise so as to consider it to be reliable.The hypothesis must be specific and should have scope for conducting more tests.

4. The way of explanation of hypothesis must be very simple and it should also be understood that simplicity of hypothesis is not related to its significance.

5. There are 6 types of hypotheses out of which two of them are used in statistical tests
> 5.1 **Null Hypothesis(H0)**: It provides the statement which is contrary to hypothesis. Itâ€™s a negative statement, and there is no relationship between independent and dependent variable. The symbol is denoted by â€œHOâ€.<br>
> 5.2 **Alternative Hypothesis(Ha)**: The alternate hypothesis is just an alternative to the null hypothesis.

6. The statistical tests are divided into two main categories. Below table describes the differences

**5.2 Performing Statistical Tests on columns**

**Mann-Whitney U-Test**

**Age vs Attriton:**

**Null Hypothesis(H0):** There is no significant difference in age of employees who left the company(attrition is Yes) to age of employees who stayed in company(attrition is No)<br>
**Alternative Hypothesis(Ha):** There is a significant difference in age of employees who left the company(attrition is Yes) to age of employees who stayed in company(attrition is No)

```
#Import scipy package for statisitical learning
from scipy.stats import mannwhitneyu

#Divide the dataset into two parts, one dataset has attrition as yes and another as attrition as no
data_yes = dataset[dataset['Attrition'] == "Yes"]
data_no = dataset[dataset['Attrition'] == "No"]

#Peroform Mann-Whitney utest on age vs attrition
stats, p = mannwhitneyu(data_yes['Age'], data_no['Age'])
print("P- value is: ", p)

if p>0.05:
  print("H0 is accepted!")
else:
  print("Ha is accepted")
```

From the above results, the p-value is less than critical value(0.05). So we fail the null hypothesis and accept alternative hypothesis.

**One-Way ANOVA or One-sample F-Test**<br>

The one-way analysis of variance (ANOVA) is used to determine whether there are any statistically significant differences between the means of two or more independent (unrelated) groups (although you tend to only see it used when there are a minimum of three, rather than two groups). <br>

Now we perform the one-way anova on age and attrition columns of the dataset
